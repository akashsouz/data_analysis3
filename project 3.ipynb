```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('medical_examination.csv')
df.info()
```

```python
df
```

```python
df['overweight'] = np.nan
df
```

```python
df['overweight']
```

```python
df['overweight'] = df['weight'] / ((0.01 * df['height']) ** 2)
df['overweight'] = df['overweight'].apply(lambda x: 1 if x > 25 else 0)
df
```

```python
df
```

```python
df['gluc'] = df['gluc'].apply(lambda x: 1 if x > 1 else 0)
df['cholesterol'] = df['cholesterol'].apply(lambda x: 1 if x > 1 else 0)
df
```

```python
df
```

```python
df_cat = pd.melt(df, id_vars=['age','cardio'], value_vars=['active', 'alco', 'cholesterol', 'gluc', 'overweight', 'smoke'])
```

```python
df_cat 
```

```python
fig = sns.catplot(x="variable", col="cardio", col_wrap=2, 
                data=df_cat, 
                kind="count", height=7, aspect=1, hue = "value")
fig.set_axis_labels("", "total")
```

```python
df_heat = df
```

```python
df_heat = df_heat.drop(df_heat[df_heat['ap_lo'] > df_heat['ap_hi']].index)
df_heat
```

```python
df_heat = df_heat.drop(df_heat[df_heat['height'] < df_heat['height'].quantile(0.025)].index)
df_heat
```

```python
df_heat = df_heat.drop(df_heat[df_heat['height'] > df_heat['height'].quantile(0.975)].index)
df_heat = df_heat.drop(df_heat[df_heat['weight'] > df_heat['weight'].quantile(0.975)].index)
df_heat = df_heat.drop(df_heat[df_heat['weight'] < df_heat['weight'].quantile(0.025)].index)
df_heat
```

```python
df_heat
```

```python
mask = np.triu(np.zeros_like(df_heat.corr())) 
mask[np.triu_indices_from(mask)] = True
```

```python
plt.figure(figsize= (10, 6))
sns.heatmap(df_heat.corr(), annot=True, fmt=".1f", mask=mask, linewidth=.1)
plt.show()

```
